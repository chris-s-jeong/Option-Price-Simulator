## European Call and Put Option

Compute the price of a European call (put) option with a strike price of 110 (90) that pays at maturity:

$$P_{Call} = max(S_T - 110, 0)$$
$$P_{Put} = max(90 - S_T, 0)$$


```python
Methodology.europeanCallPrice(sim)
Methodology.europeanPutPrice(sim)
```

## Floating Lookback Call and Put Option

Compute the price of the floating lookback call and put option that pays at maturity:

$$ P_{Call} = S_T - S_{min} $$
$$ P_{Put} = S_{max} - S_T $$

$S_{min}$: Minimum price of the stock in each path

$S_{max}$: Maximum price of the stock in each path



```python
Methodology.lookBackCallPrice(sim)
Methodology.lookBackPutPrice(sim)
```

$$ P_{Call} = max(S_T - \bar{S}) $$

$\bar{S}$: Average price of the stock in each path

The function calls for the options are from the package Methodology.py. The package includes functions for the calculation of a stock price path over 0.75 years. The dynamics of the
underlying are generated by $$
x_n \Delta t = x_{n-1} \Delta t + \left( \mu - \frac{1}{2} \sigma^2 \right) \Delta t + \sigma \Delta t \epsilon_n
$$
which is a stochastic process based on Brownian motion where $\epsilon$ ~ N(0,1) 
Such a process is repeated 10,000 times and the payoffs for each option are calculated with each iteration using the payoff functions shown in the notebook.

The final price of each option is then calculated via the following:
$$
P = \left( \frac{1}{10,000} \sum_{k=1}^{10,000} X(k) \right) e^{-rT}
$$
